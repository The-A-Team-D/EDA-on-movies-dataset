```python
deleted_column_list = ['popularity','homepage','id','imdb_id','keywords','overview','production_companies','vote_count','vote_average','budget_adj', 'revenue_adj']


movie_df.drop(deleted_column_list, axis = 1, inplace = True)

movie_df.head(3)
```

```python
movie_df.shape
```

```python
movie_df.drop_duplicates(keep = 'first', inplace = True)
movie_df.shape
```

```python
col = ['budget','revenue']
movie_df[col] = movie_df[col].replace(0, np.NaN)
movie_df.dropna(subset = col, inplace = True)
```

```python
movie_df.head(3)
```

```python
movie_df.dtypes
```

```python
movie_df.release_date = pd.to_datetime(movie_df['release_date'])
```

```python
movie_df.head()
```

```python
change_type = ['budget','revenue']
movie_df[change_type] = movie_df[change_type].astype('int64')
```

```python
movie_df.head()
```

```python
movie_df.shape
```

```python
movie_df = movie_df.reset_index()
movie_df
```

```python
col = list(movie_df.columns)
col.remove('index')
movie_df = movie_df[col]
movie_df
```

## Ques 2 : What is the average number of words in movie titles between the year 2000-2005 ?

```python
movie_title = movie_df[(movie_df['release_year'] >=2000) & (movie_df['release_year'] <=2005)]
total_words = []
for i in movie_title['original_title']:
    for j in i.split(' '):
        total_words.append(j)
    
print(len(total_words))
```

```python
movie_title.shape
```

###  Answer : Average no. of words in movie title is ...

```python
Avg_no_of_words = int(len(total_words)/movie_title.shape[0])
Avg_no_of_words
```
